```python
import numpy as np
import matplotlib.pyplot as plt
import os.path

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

from tqdm import tqdm
```

```python
# Generate data
X, y = make_classification(n_samples=10000, n_features=100, n_informative=100, n_redundant=0,\
                           n_classes=10, n_clusters_per_class=1, class_sep=1.2, flip_y=0, random_state=10)

y = y.astype(int)

# Create folder
if not os.path.exists('data/'):
    os.mkdir('data/')

np.save('data/synthetic-data.npy', X)
np.save('data/synthetic-label.npy', y)
```

```python
colours = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple',\
           'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan']

plt.figure(figsize=(15,10))

for i in range(10):

    Xi = X[np.argwhere(y==i)]
    
    Xi_mean = np.mean(Xi,axis=0)[0]
    Xi_std  = np.std(Xi, axis=0)[0]

    nx = np.arange(len(Xi_mean))

    plt.subplot(5,2,i+1)
    plt.plot(Xi_mean, c=colours[i], linewidth=2)
    plt.fill_between(nx, Xi_mean-Xi_std, Xi_mean+Xi_std, facecolor=colours[i], alpha=0.2)
    plt.title('Class {}'.format(i))

plt.tight_layout()
plt.savefig('data/synthetic-data.pdf', dpi=150)
plt.show()
```
